```python
import matplotlib.pyplot as plt
import numpy as np
```

## Single-scenario simulation

```python
x = [10,100,1000]
y_10 = [4,2,2]
y_100 = [2,1,1]
y_1000 = [1,1,1]

plt.plot(range(len(x)), y_10,  marker='o',label='10 hash functions')
plt.plot(range(len(x)), y_100,  marker='o',label='100 hash functions')
plt.plot(range(len(x)), y_1000,  marker='o',label='1000 hash functions')
plt.xticks(range(len(x)),x)
plt.yticks(np.arange(1,5, 1))
plt.xlabel('Nymber of buckets')
plt.ylabel('FP')
plt.title("Single-scenario - FPs per number of buckets and hash functions") 
plt.legend(loc="upper right")


plt.show()
```

## Mutli-scenario simulation

```python
x = [10,100,1000]
y_10 = [13,10,10]
y_100 = [10,8,8]
y_1000 = [9,8,8]

plt.plot(range(len(x)), y_10,  marker='o',label='10 hash functions')
plt.plot(range(len(x)), y_100,  marker='o',label='100 hash functions')
plt.plot(range(len(x)), y_1000,  marker='o',label='1000 hash functions')
plt.xticks(range(len(x)),x)
plt.yticks(np.arange(8,14, 1))
plt.xlabel('Nymber of buckets')
plt.ylabel('FP')
plt.title("Multi-scenario - FPs per number of buckets and hash functions") 
plt.legend(loc="upper right")


plt.show()
# plt.tight_layout()
# plt.savefig('tuning_single.png')
```
